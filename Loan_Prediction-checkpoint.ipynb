Handling null values

```python
loan_df.isnull().sum()
```

```python
loan_df["Property_Area"].value_counts()
```

```python
mode1=loan_df["Gender"].mode().iloc[0]
mode2=loan_df["Dependents"].mode().iloc[0]
mode3=loan_df["Married"].mode().iloc[0]
mode5=loan_df["Self_Employed"].mode().iloc[0]
median1=loan_df["LoanAmount"].median()
median2=loan_df["Loan_Amount_Term"].median()
median3=loan_df["Credit_History"].median()
```

```python
loan_df["Gender"].fillna(mode1,inplace=True)
loan_df["Dependents"].fillna(mode2,inplace=True)
loan_df["Married"].fillna(mode3,inplace=True)
loan_df["Self_Employed"].fillna(mode5,inplace=True)
loan_df["LoanAmount"].fillna(median1,inplace=True)
loan_df["Loan_Amount_Term"].fillna(median2,inplace=True)
loan_df["Credit_History"].fillna(median3,inplace=True)
```

```python
loan_df.isnull().sum()
```

```python
loan_df.dtypes
```

dropping unwanted columns

```python
loan_df.drop(['Loan_ID'],axis=1,inplace=True)
```

```python
loan_df['Dependents'].replace(['3+'],4,inplace=True)
loan_df['Dependents'].replace(['3'],3,inplace=True)
loan_df['Dependents'].replace(['2'],2,inplace=True)
loan_df['Dependents'].replace(['1'],1,inplace=True)
loan_df['Dependents']=loan_df['Dependents'].astype(float)
```

```python
loan_df.dtypes
```

onehot encoding

```python
l_df=pd.get_dummies(loan_df,columns=['Gender','Married','Education','Self_Employed','Property_Area'])
l_df
```

```python
l_df.dtypes
```

handling and removing outliers

```python
l_df['zscore1']=(l_df['LoanAmount']-l_df['LoanAmount'].mean())/l_df['LoanAmount'].std()
l_df.drop(l_df[np.abs(l_df['zscore1'])>3].index, inplace = True)
l_df['zscore2']=(l_df['CoapplicantIncome']-l_df['CoapplicantIncome'].mean())/l_df['CoapplicantIncome'].std()
l_df.drop(l_df[np.abs(l_df['zscore2'])>3].index, inplace = True)
```

```python
l_df.drop(['zscore1','zscore2'],inplace=True,axis=1)
```

```python
x=l_df.drop(['Loan_Status'],axis=1)
y=l_df[['Loan_Status']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

logistic regression

```python
logistic_regression=LogisticRegression()
logistic_regression.fit(x_train,y_train)
print(logistic_regression.score(x_train,y_train),logistic_regression.score(x_test,y_test))
```

metrics

```python
predictions=logistic_regression.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

KNN MODEL

```python
knn=KNeighborsClassifier(n_neighbors=80,p=2) #p=1, manhattan p=2, euclidian
knn.fit(x_train,y_train)
print(knn.score(x_train,y_train),knn.score(x_test,y_test))
```

```python
predictions=knn.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

SVC MODEL

```python
clf=SVC(kernel='linear',C=2,gamma='auto')
clf.fit(x_train,y_train)
print(clf.score(x_train,y_train),clf.score(x_test,y_test))
```

```python
predictions=clf.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

ENSEMBLING MODELS - ADABOOSTING

```python
adb=AdaBoostClassifier()
adb.fit(x_train,y_train)
print(adb.score(x_train,y_train),adb.score(x_test,y_test))
```

```python
predictions=adb.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

GRADIENTBOOSTING

```python
gbc=GradientBoostingClassifier(n_estimators=30)
gbc.fit(x_train,y_train)
print(gbc.score(x_train,y_train),gbc.score(x_test,y_test))
```

```python
predictions=gbc.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

DESICION TREE

```python
dtm=DecisionTreeClassifier(max_depth=0.5,criterion='entropy')
dtm.fit(x_train,y_train)
print(dtm.score(x_train,y_train),dtm.score(x_test,y_test))
```

```python
predictions=dtm.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

RANDOM FOREST

```python
rfc=RandomForestClassifier(n_estimators=40,max_features=9)
rfc.fit(x_train,y_train)
print(rfc.score(x_train,y_train),rfc.score(x_test,y_test))
```

```python
predictions=rfc.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

BAGGING

```python
bgc1=BaggingClassifier(n_estimators=12,base_estimator=dtm)
bgc1.fit(x_train,y_train)
print(bgc1.score(x_train,y_train),bgc1.score(x_test,y_test))
```

```python
predictions=bgc1.predict(x_test)
print(metrics.classification_report(y_test,predictions))
```

AdaBoosting Classifier is suitable model since the data fits for the model accurately with accurate train and test score
